```python
# File Loading from csvv data set
data=pd.read_csv(r"C:\Users\dheer\Downloads\data.csv")
data
```

```python
# checking NULL values
data.isnull().sum()
```

```python
# Dropping Dupicate values
data.drop_duplicates(inplace=True)
```

```python
#Encoding Location Attribute

data= pd.get_dummies(data, columns=['location'], drop_first=True)
```

```python
# co relation between features
correlation=data.corr()
correlation
```

```python
# constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
# Splitting Data as Input Data and Target
X = data.drop('price', axis=1)
y = data['price']

# Splitting Data for Model Training and Model testing (Train Data: 80%, Test Data:20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Model_Training
model = LinearRegression()
model.fit(X_train, y_train)
data
```

```python
# Accuracy for prediction on train data
training_data_prediction = model.predict(X_train)
 
```

```python
score_1=metrics.r2_score(y_train,training_data_prediction)

score_2=metrics.mean_absolute_error(y_train,training_data_prediction)

print("Root_mean_square:",score_1)
print("absolute_mean_square:",score_2)
```

```python
#Actual Prediction
test_data_prediction=model.predict(X_test)
```

```python
#Evaluation Metrics
score_1=metrics.r2_score(y_test,test_data_prediction)

score_1=metrics.mean_absolute_error(y_test,test_data_prediction)

print("Root_mean_square:",score_1)
print("absolute_mean_square:",score_2)
```

```python
#Scatter PLot of Prediction vs Actual Prices
plt.scatter(y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```
